```python
input_str = input().split('+')

numbers = []
for num in input_str:
    numbers.append(int(num))

numbers.sort()

str_numbers = []
for nums in numbers:
    str_numbers.append(str(nums))

result = '+'.join(str_numbers)
print(result)
```

```python
s = input().split('+')
s = [int(num) for num in s]
s.sort()
result = '+'.join(map(str,s))
print(result)
```
