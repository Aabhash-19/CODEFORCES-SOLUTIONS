```python
n = int(input())
lucky_numbers = []

for num in range(4, 1001):
    num_str = str(num)
    all_digits_lucky = True
    
    for digit in num_str:
        if digit != '4' and digit != '7':
            all_digits_lucky = False
            break
    
    if all_digits_lucky:
        lucky_numbers.append(num)

almost_lucky = False

for lucky in lucky_numbers:
    if n % lucky == 0:
        almost_lucky = True
        break  

if almost_lucky:
    print('YES')
else:
    print('NO')
```
