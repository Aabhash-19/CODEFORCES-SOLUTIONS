```python
n = int(input())
heights = list(map(int, input().split()))

max_pos = 0
min_pos = 0

for i in range(n):
    if heights[i] > heights[max_pos]:
        max_pos = i
    if heights[i] <= heights[min_pos]:
        min_pos = i

if max_pos < min_pos:
    result = max_pos + (n - 1 - min_pos)
else:
    result = max_pos + (n - 1 - min_pos) - 1

print(result)
```
