```python
s = input()

if s.isupper() or (s[0].islower() and s[1:].isupper()):
    flipped = ''
    for char in s:
        if char.isupper():
            flipped += char.lower()  
        else:
            flipped += char.upper()
    print(flipped)
elif len(s) == 1 and s.islower():
    print(s.upper())    
else:
    print(s)
```
