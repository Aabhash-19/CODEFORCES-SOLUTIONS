```python
n = int(input())
heights = list(map(int, input().split()))

max_water = 0

for i in range(n):
    count = 1

    left = i - 1
    while left >= 0 and heights[left] <= heights[left + 1]:
        count += 1
        left -= 1

    right = i + 1
    while right < n and heights[right] <= heights[right - 1]:
        count += 1
        right += 1

    if count > max_water:
        max_water = count

print(max_water)        
```
