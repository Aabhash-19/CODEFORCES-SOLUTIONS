```python
matrix = [list(map(int, input().split())) for _ in range(5)]
found = False
for i in range(5):
    for j in range(5):
        if matrix[i][j] == 1:
            rows, cols = i+1, j+1
            moves = abs(rows-3) + abs(cols-3) ## Manhattan Distance
            print(moves)
            found = True
            break
    if found:
        break    
```
