```python
n1, n2 = map(int, input().split())
primeNumbers = []

for num in range(2, 100):  
    isPrime = True
    for i in range(2, int(num**0.5) + 1):  
        if num % i == 0:
            isPrime = False
            break
    if isPrime:
        primeNumbers.append(num)  

if n1 in primeNumbers:
    index = primeNumbers.index(n1)
    if index + 1 < len(primeNumbers) and primeNumbers[index + 1] == n2:
        print('YES')    
    else:
        print('NO')
else:
    print('NO')        
```
