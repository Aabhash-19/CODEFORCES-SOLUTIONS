```python
n, t = map(int, input().split())
q = list(input().upper())

while(t!=0):
    index_swap = []
    for i in range(n-1):
        if q[i]=="B" and q[i+1]=="G":
            index_swap.append(i)

    for i in index_swap:
        q[i], q[i+1] = q[i+1], q[i]

    t-=1

for i in q:
    print(i, end="")    
```
